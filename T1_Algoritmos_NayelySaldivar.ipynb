En orden de comenzar a comparar algoritmos y su eficiencia, nuestro primer acercamiento será comparar y analizar algunos órdenes de crecimiento establecidos.


```
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import math

def graficar(lista1,lista2):
  plt.plot(lista1)   # Dibuja el gráfico
  plt.xlabel("n")   # Inserta el título del eje X 
  plt.ylabel("f(n)")   # Inserta el título del eje Y
  plt.ioff()   # Desactiva modo interactivo de dibujo
  plt.plot(lista2,marker = "o")   # No dibuja datos de lista2
  plt.ion()   # Activa modo interactivo de dibujo
  plt.plot(lista2,marker = "*")   # Dibuja datos de lista2 sin borrar datos de lista1
  return print("Listo!")
```

```
rango=[10,100,1000,10000]
```

```
# Ejercicio 1
# O(1) vs O(logn)
lista1=[]
lista2=[]

for i in range(1,5):
  lista1.append(1) 
  lista2.append(np.log(i))
graficar(lista1,lista2)
```

```
# Tabla
simulacion={}
for r in rango:
  simulacion[r]=(1,np.log(r))

simulacion
```

Podemos ver que para valores de n cercanos a 2 por la izquierda f=1 es T(n) por f=log(n)



```
# Ejercicio 2
# O(n) vs O(nlogn)
lista1=[]
lista2=[]
for i in range(1,8):
  lista1.append(i) 
  lista2.append(np.log(i)*i)
graficar(lista1,lista2)
```

```
# Tabla
simulacion={}
for r in rango:
  simulacion[r]=(r,np.log(r)*r)

simulacion
```

Similar al caso anterior, tenemos que para valores de n cercanos a 2 por la izquierda f=n es T(n) por f=nlog(n)


```
# Ejercicio 3
# O(n**2) vs O(n**3)

lista1=[]
lista2=[]

for i in range(1,6):
  lista1.append(i**2) 
  lista2.append(i**3)
graficar(lista1,lista2)
```

```
# Tabla
simulacion={}
for r in rango:
  simulacion[r]=(r**2,r**3)

simulacion
```

En este ejemplo, se ve muy marcado que f=n^2 es T(n) para f=n^3

```
# Ejercicio 4
# a**n vs n!

# Para este ejemplo tomemmos a=2
# definimos el cálculo del factorial
def fact(n):
  if n==0 or n ==1:
    return 1
  else: 
    for i in range(2,n):
      n*=i
    return n

lista1=[]
lista2=[]

for i in range(5):
  lista1.append(2**i) 
  lista2.append(fact(i))
graficar(lista1,lista2)
```

```
# Tabla
simulacion={}
for r in rango:
  simulacion[r]=(2**r,fact(r))

simulacion
```

Para este caso, f=n! es T(n) para f=2^n con n>3.25 aproximadamente

```
# Ejercicio 5
# n! vs n**n

lista1=[]
lista2=[]

for i in range(4):
  lista1.append(fact(i)) 
  lista2.append(i**i)
graficar(lista1,lista2)
```

```
# Tabla
simulacion={}
for r in rango:
  simulacion[r]=(fact(r),r**r)

simulacion
```

En este caso decidí dejar pequeño el rango ya que es muy evidente que f=n! es T(n) para f=n^n con n>1
